```python
import pandas as pd
import psycopg2

# 1. Konfigurasi Koneksi ke Warehouse
db_config = {
    "host": "postgres-warehouse",
    "port": "5432",
    "database": "f1_datawarehouse",
    "user": "warehouse_user",
    "password": "warehouse_password"
}
conn = psycopg2.connect(**db_config)

# 2. Cek Jumlah Data di Tabel Baru 'lap_telemetry_full'
try:
    query = "SELECT count(*) as total_rows FROM lap_telemetry_full"
    df_count = pd.read_sql(query, conn)
    
    total_data = df_count['total_rows'][0]
    print(f"\nTOTAL DATA: {total_data:,} Baris")
    
    # Validasi
    if total_data > 50000:
        print("Data detail (Lap-Level) sudah tersimpan.")
    else:
        print("Jumlah data sedikit. Cek apakah filter terlalu ketat.")

    # 3. Intip 5 Data Teratas
    print("\n5 DATA TERATAS")
    df_sample = pd.read_sql("SELECT * FROM lap_telemetry_full LIMIT 5", conn)
    display(df_sample)

except Exception as e:
    print(f"Error: {e}")
    print("Kemungkinan tabel 'lap_telemetry_full' belum terbentuk.")

finally:
    conn.close()
```
